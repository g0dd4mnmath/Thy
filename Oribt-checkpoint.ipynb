# 08:EulerMethod

dx=v;
dv = - GM/r^2 = - GM/(x^2+y^2) = a;
GM = 4*pi^2;


## funcation

```python
GM = 4*math.pi*math.pi
a = 1  # 椭圆 焦点在x轴上
e = 0  # 椭圆偏心率


def eulerMehod(h, step):
    label = 'dt=%s' % h
    x_var = [0]
    y_var = [a*(1-e)]
    u_var = [-math.sqrt((GM/a)*((1+e)/(1-e)))]  # x方向
    v_var = [0]  # y方向
    a_var = [-GM/(a*(1-e)*a*(1-e))]
    for i in np.arange(0, step):
        temp = math.sqrt(x_var[-1]*x_var[-1] + y_var[-1]*y_var[-1])  # r
        tempX = x_var[-1]/temp  # 与x轴的夹角
        tempY = y_var[-1]/temp  # 与y轴的夹角
        x_var.append(x_var[-1]+h*u_var[-1])
        y_var.append(y_var[-1]+h*v_var[-1])
        u_var.append(u_var[-1]+h*a_var[-1]*tempX)
        v_var.append(v_var[-1]+h*a_var[-1]*tempY)
        a_var.append(-GM/(x_var[-1]*x_var[-1] + y_var[-1]*y_var[-1]))
    fPlus = [x_var, y_var, label]
    return fPlus


def eulerCromer(h, step):
    #label = 'dt=%s' % h
    label = 'Euler-Cromer'
    x_var = [0]
    y_var = [a*(1-e)]
    u_var = [-math.sqrt((GM/a)*((1+e)/(1-e)))]
    v_var = [0]
    a_var = [-GM/(a*(1-e)*a*(1-e))]
    e_var = [-GM/(a*(1-e))+(1/2)*(u_var[-1]*u_var[-1]+v_var[-1]*v_var[-1])]
    t_var = [0]
    for i in np.arange(0, step):
        temp = math.sqrt(x_var[-1]*x_var[-1] + y_var[-1]*y_var[-1])
        tempX = x_var[-1]/temp
        tempY = y_var[-1]/temp
        u_var.append(u_var[-1]+h*a_var[-1]*tempX)  # 先改变v,再改变x
        v_var.append(v_var[-1]+h*a_var[-1]*tempY)
        x_var.append(x_var[-1]+h*u_var[-1])
        y_var.append(y_var[-1]+h*v_var[-1])
        a_var.append(-GM/(x_var[-1]*x_var[-1] + y_var[-1]*y_var[-1]))
        e_var.append(-GM/(math.sqrt(x_var[-1]*x_var[-1] + y_var[-1]*y_var[-1]))+(1/2)*(u_var[-1]*u_var[-1]+v_var[-1]*v_var[-1]))
        t_var.append(t_var[-1]+h)
    fcromer = [x_var, y_var, label, t_var, e_var]
    return fcromer


def circle(r):
    a = 0
    b = 0  # 圆心坐标
    theta = np.arange(0, 2*np.pi, 0.01)
    x = a + r * np.cos(theta)
    y = b + r * np.sin(theta)
    return[x, y]
```

## figuer

```python
cir = circle(1)
dt0 = eulerMehod(h = 0.1,step = 10)
dt1 = eulerMehod(0.05,20)
dt2 = eulerMehod(0.025,40)
dt3 = eulerMehod(0.0125,80)
ec = eulerCromer(0.0125,80)

fig = plt.figure(figsize=(9,6))
plt.scatter(0,0)
plt.plot(cir[0], cir[1], '--k')
plt.plot(dt0[0], dt0[1], 'k', label=dt0[2])
plt.plot(dt1[0], dt1[1], 'b', label=dt1[2])
plt.plot(dt2[0], dt2[1], 'r', label=dt2[2])
plt.plot(dt3[0], dt3[1], 'g', label=dt3[2])
plt.plot(ec[0],ec[1],'m',label='Euler-Cromer method')
plt.xlim(-3,3)
plt.ylim(-2,2)
plt.legend()
plt.title('Orbits: Euler\'s Method')
plt.savefig('EulerMethod.jpeg',dpi=200)
```

## funcation

```python
GM = 4*math.pi*math.pi
a = 1  # 椭圆 焦点在x轴上
e = 0  # 椭圆偏心率


def secondRoungKutte(h, step):
    label = 'R-K 2nd oder'
    t_var = [0]
    x_var = [0]
    y_var = [a*(1-e)]
    u_var = [-math.sqrt((GM/a)*((1+e)/(1-e)))]
    v_var = [0]
    a_var = [-GM/(a*(1-e)*a*(1-e))]
    e_var = [-GM/(math.sqrt(x_var[-1]*x_var[-1] + y_var[-1]*y_var[-1]))+(1/2)*(u_var[-1]*u_var[-1]+v_var[-1]*v_var[-1])]
    for i in np.arange(0, step):
        temp = math.sqrt(x_var[-1]*x_var[-1] + y_var[-1]*y_var[-1])  # r
        tempX = x_var[-1]/temp  # 与x轴的夹角
        tempY = y_var[-1]/temp  # 与y轴的夹角

        xTemp = x_var[-1]+(h/2)*u_var[-1]
        yTemp = y_var[-1]+(h/2)*v_var[-1]
        uTemp = u_var[-1]+(h/2)*a_var[-1]*tempX
        vTemp = v_var[-1]+(h/2)*a_var[-1]*tempY

        temp = math.sqrt(xTemp*xTemp + yTemp*yTemp)     # 更新r,夹角
        tempX= xTemp/temp
        tempY= yTemp/temp
        aTemp = -GM/(xTemp*xTemp + yTemp*yTemp)

        x_var.append(x_var[-1]+h*uTemp)
        y_var.append(y_var[-1]+h*vTemp)
        u_var.append(u_var[-1]+h*aTemp*tempX)
        v_var.append(v_var[-1]+h*aTemp*tempY)
        a_var.append(-GM/(x_var[-1]*x_var[-1] + y_var[-1]*y_var[-1]))
        e_var.append(-GM/(math.sqrt(x_var[-1]*x_var[-1] + y_var[-1]*y_var[-1]))+(1/2)*(u_var[-1]*u_var[-1]+v_var[-1]*v_var[-1]))
        t_var.append(t_var[-1]+h)
    fRungeKutte = [x_var, y_var, label, t_var, e_var]
    return fRungeKutte
```

```python
a = 1  # 椭圆 焦点在x轴上
e = 0  # 椭圆偏心率

#dx dv ddx = du; ddy = dv 
def dxyuv(x,y,u,v):
    dx = u
    dy = v
    ddX = - GM*x*math.pow(x*x+y*y, -3/2)
    ddY = - GM*y*math.pow(x*x+y*y, -3/2)
    return [dx,dy,ddX,ddY]

def fourthRungeKutte(h, step):
    label = 'R-K 4th oder'
    t_var = [0]
    x_var = [0]
    y_var = [a*(1-e)]
    u_var = [-math.sqrt((GM/a)*((1+e)/(1-e)))]
    v_var = [0]
    a_var = [-GM/(a*(1-e)*a*(1-e))]
    e_var = [-GM/(math.sqrt(x_var[-1]*x_var[-1] + y_var[-1]*y_var[-1]))+(1/2)*(u_var[-1]*u_var[-1]+v_var[-1]*v_var[-1])]
    for i in np.arange(0, step):
        k1 = dxyuv(x=x_var[-1], y=y_var[-1], u=u_var[-1],v=v_var[-1])
        k2 = dxyuv(x_var[-1]+(h/2)*k1[0], y_var[-1]+(h/2)*k1[1], u_var[-1]+(h/2)*k1[2], v_var[-1]+(h/2)*k1[3])
        k3 = dxyuv(x_var[-1]+(h/2)*k2[0], y_var[-1]+(h/2)*k2[1], u_var[-1]+(h/2)*k2[2], v_var[-1]+(h/2)*k2[3])
        k4 = dxyuv(x_var[-1]+h*k3[0], y_var[-1]+h*k3[1], u_var[-1]+h*k3[2], v_var[-1]+h*k3[3])
        x_var.append(x_var[-1]+(h/6)*(k1[0]+2*k2[0]+2*k3[0]+k4[0]))
        y_var.append(y_var[-1]+(h/6)*(k1[1]+2*k2[1]+2*k3[1]+k4[1]))
        u_var.append(u_var[-1]+(h/6)*(k1[2]+2*k2[2]+2*k3[2]+k4[2]))
        v_var.append(v_var[-1]+(h/6)*(k1[3]+2*k2[3]+2*k3[3]+k4[3]))
        a_var.append(-GM/(x_var[-1]*x_var[-1] + y_var[-1]*y_var[-1]))
        e_var.append(-GM/(math.sqrt(x_var[-1]*x_var[-1] + y_var[-1]*y_var[-1]))+(1/2)*(u_var[-1]*u_var[-1]+v_var[-1]*v_var[-1]))
        t_var.append(t_var[-1]+h)
    fRungeKutte = [x_var, y_var, label, t_var, e_var]
    return fRungeKutte

```

### Runge-Kutta-Fehlberg

```python
a = 1  # 椭圆 焦点在x轴上
e = 0  # 椭圆偏心率

#dx dv ddx = du; ddy = dv 
def dxyuv(x,y,u,v):
    dx = u
    dy = v
    ddX = - GM*x*math.pow(x*x+y*y, -3/2)
    ddY = - GM*y*math.pow(x*x+y*y, -3/2)
    return [dx,dy,ddX,ddY]

def rungeKuttaFehlberg(h, step,T):
    label = 'RungeKuttaFehlberg'
    x_var = [0]
    y_var = [a*(1-e)]
    u_var = [-math.sqrt((GM/a)*((1+e)/(1-e)))]
    v_var = [0]
    a_var = [-GM/(a*(1-e)*a*(1-e))]
    h0 = [0]
    t0 = [0]
    testh = 1
    reset = 0

    for i in np.arange(0, step):
        while True:
            k1 = dxyuv(x=x_var[-1], y=y_var[-1], u=u_var[-1],v=v_var[-1])
            k2 = dxyuv(x_var[-1]+(h/4)*k1[0], y_var[-1]+(h/4)*k1[1], u_var[-1]+(h/4)*k1[2], v_var[-1]+(h/4)*k1[3])
            k3 = dxyuv(x_var[-1]+(3/32)*h*k1[0]+(9/32)*h*k2[0], y_var[-1]+(3/32)*h*k1[1]+(9/32)*h*k2[1], u_var[-1]+(3/32)*h*k1[2]+(9/32)*h*k2[2], v_var[-1]+(3/32)*h*k1[3]+(9/32)*h*k2[3])
            k4 = dxyuv(x_var[-1]+(1932/2197)*h*k1[0]-(7200/2197)*h*k2[0]+(7296/2197)*h*k3[0], y_var[-1]+(1932/2197)*h*k1[1]-(7200/2197)*h*k2[1]+(7296/2197)*h*k3[1], u_var[-1]+(1932/2197)*h*k1[2]-(7200/2197)*h*k2[2]+(7296/2197)*h*k3[2], v_var[-1]+(1932/2197)*h*k1[3]-(7200/2197)*h*k2[3]+(7296/2197)*h*k3[3])
            k5 = dxyuv(x_var[-1]+(439/216)*h*k1[0]-8*h*k2[0]+(3680/513)*h*k3[0]-(845/4104)*h*k4[0], y_var[-1]+(439/216)*h*k1[1]-8*h*k2[1]+(3680/513)*h*k3[1]-(845/4104)*h*k4[1], u_var[-1]+(439/216)*h*k1[2]-8*h*k2[2]+(3680/513)*h*k3[2]-(845/4104)*h*k4[2], v_var[-1]+(439/216)*h*k1[3]-8*h*k2[3]+(3680/513)*h*k3[3]-(845/4104)*h*k4[3])
            k6 = dxyuv(x_var[-1]-(8/27)*h*k1[0]+2*h*k2[0]-(3544/2565)*h*k3[0]+(1859/4104)*h*k4[0]-(11/40)*h*k5[0], y_var[-1]-(8/27)*h*k1[1]+2*h*k2[1]-(3544/2565)*h*k3[1]+(1859/4104)*h*k4[1]-(11/40)*h*k5[1], u_var[-1]-(8/27)*h*k1[2]+2*h*k2[2]-(3544/2565)*h*k3[2]+(1859/4104)*h*k4[2]-(11/40)*h*k5[2], v_var[-1]-(8/27)*h*k1[3]+2*h*k2[3]-(3544/2565)*h*k3[3]+(1859/4104)*h*k4[3]-(11/40)*h*k5[3])
            x4 = x_var[-1]+h*((25/216)*k1[0]+(1408/2565)*k3[0]+(2197/4104)*k4[0]-(1/5)*k5[0])
            x5 = x_var[-1]+h*((16/135)*k1[0]+(6656/12825)*k3[0]+(28561/56430)*k4[0]-(9/50)*k5[0]+(2/55)*k6[0])
            xe = h*math.fabs(x4-x5)     #h*math.fabs((1/360)*k1[0]+(6656/12825)*k3[0]+(28561/56430)*k4[0]-(9/50)*k5[0]+(2/55)*k6[0])
            xt = T*math.fabs(x4)
            y4 = y_var[-1]+h*((25/216)*k1[1]+(1408/2565)*k3[1]+(2197/4104)*k4[1]-(1/5)*k5[1])
            y5 = y_var[-1]+h*((16/135)*k1[1]+(6656/12825)*k3[1]+(28561/56430)*k4[1]-(9/50)*k5[1]+(2/55)*k6[1])
            ye = h*math.fabs(y4-y5)     #h*math.fabs((1/360)*k1[1]+(6656/12825)*k3[1]+(28561/56430)*k4[1]-(9/50)*k5[1]+(2/55)*k6[1])
            yt = T*math.fabs(y4)
            if xe<xt and ye<yt:
                testh = 1
                x4 = x5
                y4 = y5
                u4 = u_var[-1]+h*((16/135)*k1[2]+(6656/12825)*k3[2]+(28561/56430)*k4[2]-(9/50)*k5[2]+(2/55)*k6[2])
                v4 = v_var[-1]+h*((16/135)*k1[3]+(6656/12825)*k3[3]+(28561/56430)*k4[3]-(9/50)*k5[3]+(2/55)*k6[3])
                hTemp = h
                h = 0.8*math.pow((T*math.fabs(x4)/xe), 1/5)*h
                h0.append(abs(h-hTemp))
                t0.append(t0[-1]+h)
                break
            elif testh == 1:
                h = 0.8*math.pow(((T*math.fabs(x4))/xe), 1/5)*h
                testh = 2
                reset +=1
            else:
                h = h/2
                reset +=1
        x_var.append(x4)
        y_var.append(y4)
        u_var.append(u4)
        v_var.append(v4)
        a_var.append(-GM/(x_var[-1]*x_var[-1] + y_var[-1]*y_var[-1]))
    fRungeKutte = [x_var, y_var, label,t0,h0,reset]
    return fRungeKutte

```

### half

```python
a = 1  # 椭圆 焦点在x轴上
e = 0  # 椭圆偏心率

#dx dv ddx = du; ddy = dv 
def dxyuv(x,y,u,v):
    dx = u
    dy = v
    ddX = - GM*x*math.pow(x*x+y*y, -3/2)
    ddY = - GM*y*math.pow(x*x+y*y, -3/2)
    return [dx,dy,ddX,ddY]

def kOf4thRK(x,y,u,v,h):
    k1 = dxyuv(x=x, y=y, u=u, v=v)
    k2 = dxyuv(x+(h/2)*k1[0], y+(h/2)*k1[1],u+(h/2)*k1[2], v+(h/2)*k1[3])
    k3 = dxyuv(x+(h/2)*k2[0], y+(h/2)*k2[1],u+(h/2)*k2[2], v+(h/2)*k2[3])
    k4 = dxyuv(x+h*k3[0], y+h*k3[1],u+h*k3[2], v+h*k3[3])
    xVar = x+(h/6)*(k1[0]+2*k2[0]+2*k3[0]+k4[0])
    yVar = y+(h/6)*(k1[1]+2*k2[1]+2*k3[1]+k4[1])
    uVar = u+(h/6)*(k1[2]+2*k2[2]+2*k3[2]+k4[2])
    vVar = v+(h/6)*(k1[3]+2*k2[3]+2*k3[3]+k4[3])
    return[xVar,yVar,uVar,vVar]


def adaptiveFourthRungeKutte(h, step, T,s1=0.9,s2=4.0):
    label = 'Adaptive R-K 4th oder'
    x_var = [0]
    y_var = [a*(1-e)]
    u_var = [-math.sqrt((GM/a)*((1+e)/(1-e)))]
    v_var = [0]
    hVar = [0]
    tVar = [0]
    testh = 1 ##控制只进行一次再尝试
    reset = 0 #记录尝试次数

    for i in np.arange(0, step):
        while True:
            single = kOf4thRK(x=x_var[-1], y=y_var[-1], u=u_var[-1], v=v_var[-1],h=h)
            half1 = kOf4thRK(x_var[-1],y_var[-1],u_var[-1],v_var[-1],h/2)
            half2 = kOf4thRK(half1[0],half1[1],half1[2],half1[3],h/2)
            errorRel=[math.fabs((half2[i]-single[i])/((half2[i]+single[i])/2)) for i  in range(len(single))] #[error_x,errory]
            errMax = max(errorRel)
            if errMax<T :
                testh=1
                rek = half2
                hTemp = h
                h = min(max(s1*hTemp*math.pow(T/errMax, 1/5), hTemp/s2), s2*hTemp)
                hVar.append(abs(h-hTemp))
                tVar.append(tVar[-1]+h)
                break
            elif testh==1:
                h = min(max(s1*h*math.pow(T/errMax, 1/5), h/s2), s2*h)
                testh=2
                reset+=1
            else:
                h=h/2
                reset+=1

        x_var.append(rek[0])
        y_var.append(rek[1])
        u_var.append(rek[2])
        v_var.append(rek[3])
    fRungeKutte = [x_var, y_var, label, tVar, hVar, reset]
    return fRungeKutte

```

## funcation

```python
GM = 4 * math.pi * math.pi
a = 1  # 椭圆 焦点在x轴上
e = 0  # 椭圆偏心率


def kickDriftKcik(h, step):
    label = 'velocity Verlet'
    t_var = [0]
    x_var = [0]
    y_var = [a * (1 - e)]
    u_var = [-math.sqrt((GM / a) * ((1 + e) / (1 - e)))]
    v_var = [0]
    a_var = [-GM / (a * (1 - e) * a * (1 - e))]
    e_var = [
        -GM / (math.sqrt(x_var[-1] * x_var[-1] + y_var[-1] * y_var[-1])) +
        (1 / 2) * (u_var[-1] * u_var[-1] + v_var[-1] * v_var[-1])
    ]
    for i in np.arange(0, step):
        temp = math.sqrt(x_var[-1] * x_var[-1] + y_var[-1] * y_var[-1])  # r
        tempX = x_var[-1] / temp  # 与x轴的夹角
        tempY = y_var[-1] / temp  # 与y轴的夹角

        uTemp = u_var[-1] + (h / 2) * a_var[-1] * tempX  #u_n+1/2
        vTemp = v_var[-1] + (h / 2) * a_var[-1] * tempY  #v_n+1/2

        xTemp = x_var[-1] + h * uTemp  #x_n+1
        yTemp = y_var[-1] + h * vTemp  #y_n+1

        temp = math.sqrt(xTemp * xTemp + yTemp * yTemp)  # 更新r,夹角
        tempX = xTemp / temp
        tempY = yTemp / temp
        aTemp = -GM / (xTemp * xTemp + yTemp * yTemp)

        uTemp = uTemp + (h / 2) * aTemp * tempX  # u_n+1
        vTemp = vTemp + (h / 2) * aTemp * tempY  # v_n+1

        x_var.append(xTemp)
        y_var.append(yTemp)
        u_var.append(uTemp)
        v_var.append(vTemp)
        a_var.append(-GM / (x_var[-1] * x_var[-1] + y_var[-1] * y_var[-1]))
        e_var.append(
            -GM / (math.sqrt(x_var[-1] * x_var[-1] + y_var[-1] * y_var[-1])) +
            (1 / 2) * (u_var[-1] * u_var[-1] + v_var[-1] * v_var[-1]))
        t_var.append(t_var[-1] + h)
    fkdk = [x_var, y_var, label, t_var, e_var]
    return fkdk
```
